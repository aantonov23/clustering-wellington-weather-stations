```python
import pandas as pd
import os
from dotenv import load_dotenv

load_dotenv()
BASE_DATA_PATH = os.environ.get('BASE_DATA_PATH')
```

```python
path_filled_stations = os.path.join(BASE_DATA_PATH, r'v2_station_dfs')
stations = os.listdir(path_filled_stations)
stations
```

```python
list_all_stations_raw = []
for station in stations:
    df = pd.read_csv(os.path.join(path_filled_stations, station), parse_dates=['date'])
    
    list_all_stations_raw.append(df)
    # print(type(df['Observation time UTC'][0]))
    # display(df)
display(list_all_stations_raw[15])
print(len(list_all_stations_raw))
```

```python
df_all_stations_raw = pd.concat(list_all_stations_raw, ignore_index=True)
df_all_stations_raw.info()
display(df_all_stations_raw)
```

```python
df_all_stations_raw.to_csv(os.path.join(BASE_DATA_PATH, r'v2_df_all_stations_raw.csv'), index=False)
```
