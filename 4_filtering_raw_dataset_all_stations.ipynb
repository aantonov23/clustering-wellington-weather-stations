```python
import os
import pandas as pd
from matplotlib import pyplot as plt


BASE_DATA_PATH = r'C:\D\Whitireia\ARP\data'
```

```python
df_all_stations_raw = pd.read_csv(os.path.join(BASE_DATA_PATH, r'v2_df_all_stations_raw.csv'))
df_all_stations_raw
```

```python
df = df_all_stations_raw

all_ids = df['station_id'].unique()

common_dates = set(df[df['station_id'] == all_ids[0]]['date'])
for id_ in all_ids[1:]:
    id_dates = set(df[df['station_id'] == id_]['date'])
    common_dates &= id_dates  # keep the only dates common to all ids

df_filtered = df[df['date'].isin(common_dates)]
df_filtered = df_filtered.sort_values(['station_id', 'date'], ascending=[True, False]).reset_index(drop=True)
display(df_filtered)
print(df_filtered.info())
```

```python
def split_df_by_station_id(df: pd.DataFrame) -> list:
    return list(df.groupby('station_id'))
```

```python
splits = split_df_by_station_id(df_filtered)

min_len = len(splits[0][1])
min_len_st_id = splits[0][0]
for i,v in enumerate(splits):
    print(f"{i+1} id: {v[0]};  len: {len(v[1])}")
    if min_len > len(v[1]):
        min_len = len(v[1])
        min_len_st_id = v[0]
print(f"Min rows station: {min_len_st_id}, Min len: {min_len}")
```

```python
df_filtered.to_csv(os.path.join(BASE_DATA_PATH, r'v2_df_all_stations_filtered.csv'), index=False)
```
