```python
import os
import pandas as pd

BASE_DATA_PATH = r'C:\D\Whitireia\ARP\data'
```

```python
folder_path = os.path.join(BASE_DATA_PATH, r'station_folders')
csv_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith(".csv")]
os.listdir(folder_path)
```

```python
dict_stations = {}
for file_name in os.listdir(folder_path):
    print(file_name)
    station_id = int(file_name.split('__')[0])
    file_path = os.path.join(folder_path, file_name)
    df_parameter = pd.read_csv(file_path, parse_dates=['Observation time UTC'])
    if station_id not in dict_stations:
        df_parameter['station_id'] = station_id       
        dict_stations[station_id] = df_parameter
        print(dict_stations[station_id])
    else:
        df_station = dict_stations[station_id]
        print(df_station)
        print(df_parameter)
        result_df = pd.merge(df_station, df_parameter, on='Observation time UTC', how='outer')
        result_df['station_id'] = result_df['station_id'].fillna(station_id)
        dict_stations[station_id] = result_df.sort_values('Observation time UTC', ascending=False)

display(result_df)
```

```python
len_df = []
for station_id in dict_stations:
    df = dict_stations[station_id]
    new_df = df.dropna(axis=0, ignore_index=True).round(2)
    new_df['station_id'] = new_df['station_id'].astype(int)
    new_df = new_df[['Observation time UTC', 'station_id', 'Rainfall [mm]', 'Relative humidity [%]', 'Mean Temperature [Deg C]', 'Speed [m/s]']]
    new_cols = ['date', 'station_id', 'rain', 'relative_hum', 'mean_temp', 'wind']
    new_df.columns = new_cols

    len_df.append(len(new_df))
    dict_stations[station_id] = new_df
    new_df.to_csv(os.path.join(BASE_DATA_PATH, r'v2_station_dfs', f"{station_id}.csv"), index=False)

print(f"Lenght of the dfs: {len_df}")
new_df
```
