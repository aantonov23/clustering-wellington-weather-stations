## Filling missing values at several stations

```python
import os
import pandas as pd
from functools import reduce

BASE_DATA_PATH = r'C:\D\Whitireia\ARP\data'
```

### 3145 - Wind

```python
station = 3145
station_column = 'Wind'
duration_param = 'D.csv'

df = pd.read_csv(os.path.join(BASE_DATA_PATH, r'stations_filled_all_data', f'{station}.csv'))
df['Observation time UTC'] = pd.to_datetime(df['Observation time UTC']).dt.date
df.set_index('Observation time UTC', inplace=True)
# df.drop('Observation time UTC.1', axis=1, inplace=True)
# df.to_csv(os.path.join(BASE_DATA_PATH, r'stations_filled_all_data', f'{station}.csv'))
df = df.iloc[:10950]

display(df)
```

```python
print(f"Station Id: '{station}', Station column: '{station_column}', Duration: '{duration_param}'")

temp_df = pd.read_csv(
    os.path.join(BASE_DATA_PATH, r'unzipped_data', f'{station}__{station_column}__{duration_param}'), 
    parse_dates=['Observation time UTC']
)
print(temp_df.info())
print(temp_df.sort_index(ascending=False))
# temp_df['Observation time UTC'] = pd.to_datetime(temp_df['Observation time UTC']).dt.date

temp_df.drop(
    ['Frequency [D/H]', 'Data Source', 'Direction [deg T]', 'PERIOD [hrs]'], 
    axis=1,inplace=True
)

temp_df = temp_df.groupby(pd.Grouper(key='Observation time UTC', freq='1D')).mean().sort_index(ascending=False)
# print(temp_df.info())
# display(temp_df.sort_index(ascending=False))

temp_df.index = pd.to_datetime(temp_df.index).date
print(temp_df.sort_index(ascending=False))


# temp_df.set_index('Observation time UTC', inplace=True)
# display(temp_df)
```

```python
result_df = pd.merge(df, temp_df, how='left', left_index=True, right_index=True)
result_df['Wind'] = result_df['Speed [m/s]']
result_df.drop('Speed [m/s]', axis=1, inplace=True)
```

```python
result_df.to_csv(os.path.join(BASE_DATA_PATH, r'stations_filled_all_data', f'{station}.csv'))
result_df
```

### 12442 - Wind
#### Wind data is absent for 12442
#### Data is taken as mean of two nearest stations: 8567, 3145

```python
station_result = 12442

df = pd.read_csv(os.path.join(BASE_DATA_PATH, r'stations_filled_all_data', f'{station_result}.csv'))
df['Observation time UTC'] = pd.to_datetime(df['Observation time UTC']).dt.date
df.set_index('Observation time UTC', inplace=True)
```

```python
# 8567
station = 8567

df1 = pd.read_csv(os.path.join(BASE_DATA_PATH, r'stations_filled_all_data', f'{station}.csv'))
df1['Observation time UTC'] = pd.to_datetime(df1['Observation time UTC']).dt.date
df1.set_index('Observation time UTC', inplace=True)
df1.drop(['station_id', 'Temperature',	'Screen_Observations',	'Rain'], axis=1, inplace=True)
df1.columns = ['wind_8567']
df1
```

```python
# 3145
station = 3145

df2 = pd.read_csv(os.path.join(BASE_DATA_PATH, r'stations_filled_all_data', f'{station}.csv'))
df2['Observation time UTC'] = pd.to_datetime(df2['Observation time UTC']).dt.date
df2.set_index('Observation time UTC', inplace=True)
df2.drop(['station_id', 'Temperature',	'Screen_Observations',	'Rain'], axis=1, inplace=True)
df2.columns = ['wind_3145']
df2
```

```python
data_frames = [df, df1, df2]
result_df = reduce(
    lambda left,right: pd.merge(left, right, on=['Observation time UTC'], how='outer'), data_frames
    ).sort_index(ascending=False)
result_df['Wind'] = result_df[['wind_8567', 'wind_3145']].mean(axis=1)
result_df.drop(['wind_8567', 'wind_3145'], axis=1, inplace=True)
result_df['station_id'].fillna(value=12442, inplace=True)
result_df['station_id'] = result_df['station_id'].astype(int)
result_df.to_csv(os.path.join(BASE_DATA_PATH, r'stations_filled_all_data', f'{station_result}.csv'))
result_df
```

### 31857 - Temperature

```python
# 31857__Temperature__all_data.csv
station = 31857
station_column = 'Temperature'
duration_param = 'all_data.csv'

df = pd.read_csv(os.path.join(BASE_DATA_PATH, r'stations_filled_all_data', f'{station}.csv'))
df['Observation time UTC'] = pd.to_datetime(df['Observation time UTC']).dt.date
df.set_index('Observation time UTC', inplace=True)
df
```

```python
print(f"Station Id: '{station}', Station column: '{station_column}', Duration: '{duration_param}'")

temp_df = pd.read_csv(
    os.path.join(BASE_DATA_PATH, r'unzipped_data', f'{station}__{station_column}__{duration_param}'), 
    parse_dates=['Observation time UTC']
)
# temp_df['Observation time UTC'] = pd.to_datetime(temp_df['Observation time UTC']).dt.date
# temp_df.set_index('Observation time UTC', inplace=True)

temp_df.drop(['Frequency [D/H/S]',
       'Maximum Temperature [Deg C]', 'Minimum Temperature [Deg C]',
       'Grass Temperature [Deg C]', 'Mean Relative Humidity [percent]', 'Data Source'], 
       axis=1,inplace=True
       )
temp_df = temp_df.groupby(pd.Grouper(key='Observation time UTC', freq='1D')).mean().sort_index(ascending=False)
temp_df.index = pd.to_datetime(temp_df.index).date

# temp_df = temp_df.sort_index(ascending=False)
display(temp_df)
```
